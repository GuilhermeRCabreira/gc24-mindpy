### Interpretação dos dados 


Após analisar as métricas de cada coluna, é possível compreender melhor o comportamento e as características predominantes na amostra. A seguir, apresentamos as interpretações baseadas nos valores médios e nos desvios padrão de cada variável, proporcionando uma visão mais clara sobre a frequência e a variabilidade de cada atributo analisado.

- PatientID (Identificação do Paciente) : Esta coluna serve exclusivamente para identificar de forma única cada paciente no dataset. Não possui relevância estatística direta, mas é fundamental para rastrear informações específicas dos indivíduos.

- Age (Idade): A idade média dos pacientes é de 74 anos, com um desvio padrão de aproximadamente 9 anos, o que indica uma variação considerável entre as idades. A faixa etária abrange desde pacientes com 60 anos (mínimo) até 90 anos (máximo). Isso sugere que o estudo está focado em um público idoso, faixa etária geralmente mais suscetível a doenças neurodegenerativas, como Alzheimer.

- Gender (Gênero): Esta coluna é binária, onde 0 representa masculino e 1 representa feminino. O valor médio de 0,5 sugere uma distribuição equilibrada entre os gêneros, indicando que ambos os sexos estão igualmente representados na amostra.

- Ethnicity (Etnia): Os valores desta coluna indicam as categorias de etnia dos pacientes, com 0 para Caucasian (caucasianos) e 1 para African American (afro-americanos). A média de 0,69 sugere que a maioria dos pacientes pertence às categorias 0 e 1, com predominância leve de afro-americanos. Esse dado pode ser relevante para estudar variações epidemiológicas entre diferentes grupos étnicos.

- EducaionLevel (Nível Educacional): Representa o nível educacional dos pacientes, onde níveis mais altos indicam maior escolaridade.
Os resultados indicam que, em média, os participantes completaram o ensino médio. Esse dado é relevante porque estudos mostram que níveis mais altos de escolaridade podem estar associados a uma maior reserva cognitiva, o que pode influenciar a progressão de doenças como Alzheimer.

- BMI (IMC): O IMC médio indica que a maioria dos pacientes está na categoria de sobrepeso. Há pacientes em extremos de peso, com valores mínimos indicando baixo peso (IMC de 15) e máximos chegando à obesidade grau II (IMC de 39,99). Essas condições podem influenciar negativamente a saúde geral e cognitiva dos pacientes.

- Smoking (Fumo): Esta coluna é binária, onde 0 significa não fumante e 1 significa fumante. A média de 0,28 indica que cerca de 28% dos pacientes possuem o hábito de fumar. Este dado é relevante, pois o tabagismo é um fator de risco para diversas condições de saúde, incluindo declínio cognitivo.

- AlcoholConsumption (Consumo de Álcool): O consumo médio de álcool por paciente é de 10 unidades por semana, com uma variação de 5 unidades para mais ou para menos. Isso sugere que há uma parcela considerável de pacientes com consumo moderado a alto de álcool, fator que pode impactar a saúde cerebral e o sistema nervoso.

- PhysicalActivity (Atividade Física): A prática média de 4 horas semanais indica que a maioria dos pacientes não atinge níveis de atividade física considerados ideais. A variação sugere que alguns pacientes realizam cerca de 1 hora diária de atividade física, enquanto outros são completamente sedentários. A atividade física é um fator importante para a saúde geral, especialmente em populações idosas.

- DietQuality (Qualidade da Dieta): Com uma pontuação média de 4,99, a dieta dos pacientes é classificada como razoável, com espaço significativo para melhorias. Uma alimentação equilibrada pode ter um impacto positivo na prevenção de doenças neurodegenerativas.

- SleepQuality (Qualidade do Sono): A qualidade média do sono é boa, considerando que a pontuação está acima de 7. Entretanto, a variação indica que alguns pacientes apresentam dificuldades, fator que pode estar associado ao declínio cognitivo.

- FamilyHistoryAlzheimers (Histórico Familiar de Alzheimer): Cerca de 25% dos pacientes possuem histórico familiar de Alzheimer, fator de risco importante para o desenvolvimento da doença.

- CardiovascularDisease (Doença Cardiovascular): Apenas 14% dos pacientes apresentam histórico de doenças cardiovasculares. Apesar de ser uma porcentagem baixa, é um fator relevante, pois doenças cardiovasculares podem influenciar a saúde cognitiva.

- Diabetes: Cerca de 15% dos pacientes possuem diagnóstico de diabetes, uma condição que pode estar associada a outros fatores de saúde relevantes no contexto do estudo.

- Depression (Depressão): Cerca de 20% dos pacientes apresentam depressão, o que indica uma presença significativa dessa condição entre os participantes.

- HeadInjury (Lesão na Cabeça): Apenas 9% dos pacientes relataram histórico de lesões na cabeça, representando uma minoria na amostra.

- Hypertension (Hipertensão): Apenas 14% dos pacientes possuem hipertensão, mas é um fator que deve ser monitorado, pois pode estar relacionado a complicações vasculares que afetam a saúde cerebral.

- SystolicBP (Pressão Sistólica): A média da pressão sistólica dos pacientes está acima do valor considerado ideal (<120 mmHg), o que pode indicar a necessidade de atenção em relação aos níveis de pressão arterial.

- DiastolicBP (Pressão Diastólica): Os valores médios da pressão diastólica também estão acima do limite recomendado (<80 mmHg), reforçando a relevância de monitorar essa variável.

- CholeserolTotal (Colesterol Total): O colesterol total médio está elevado, considerando que valores normais são inferiores a 200 mg/dL.

- CholesterolLDL (Colesterol LDL):Os níveis médios de LDL ("colesterol ruim") estão acima do recomendado, que é inferior a 100 mg/dL.

- CholeserolHDL (Colesterol HDL): Os níveis médios de HDL ("colesterol bom") estão dentro do intervalo ideal, geralmente acima de 40 mg/dL.

- CholeserolTriglycerides (Triglicerídeos): Os triglicerídeos apresentam uma média elevada, acima do limite recomendado de 150 mg/dL.

- MMSE (Mini-Mental State Examination): A pontuação média no MMSE sugere comprometimentos cognitivos, considerando que valores abaixo de 24 indicam possíveis alterações.

- FunctionalAssessment (Avaliação Funcional): Os escores médios indicam dificuldades moderadas nas atividades avaliadas, sugerindo impactos na funcionalidade dos pacientes.

- MemoryComplaints (Queixas de Memória): Cerca de 20% dos pacientes relatam queixas de memória, o que corresponde a uma parcela significativa.

- BehavioralProblems (Problemas Comportamentais): Apenas 15% dos pacientes relataram problemas comportamentais, indicando que essa condição não é muito frequente na amostra.

- ADL (Atividades de Vida Diária): A pontuação média indica que muitos pacientes apresentam dificuldades nas atividades de vida diária.

- Confusion (Confusão): Cerca de 20% dos pacientes apresentam confusão, o que pode indicar alguma dificuldade em processar informações ou situações.

- Disorientation (Desorientação): A desorientação foi relatada por 15% dos pacientes, representando uma parcela moderada da amostra.

- PersonalityChanges (Mudanças de Personalidade): Cerca de 15% dos pacientes relataram mudanças de personalidade, sugerindo que essa característica ocorre em menor frequência.

- DifficultyCompletingTasks (Dificuldade em Completar Tarefas): A dificuldade em completar tarefas foi identificada em 15% dos pacientes, sendo uma ocorrência moderada.

- Forgetfulness (Esquecimento): O esquecimento foi relatado por 30% dos pacientes, sendo uma das condições mais frequentes nesta amostra.

- Diagnosis (Diagnostico de Alzheimer): Cerca de 35% dos pacientes possuem diagnóstico de Alzheimer, indicando uma representatividade importante dessa condição na amostra.

```python
import numpy as np
import seaborn as sns

corrdf = df.drop("DoctorInCharge", axis=1).corr()

# Criar matriz de correlação
corr_matrix = corrdf.abs()

# Selecionar triângulo superior da matriz de correlação
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

correlacao_minima = 0.2

# Encontrar colunas que possuem alguma correlação maior que 0.2
to_select_columns = [column for column in upper.columns if any(upper[column] > correlacao_minima)]

#Encontrar linhas que possuem alguma correlação maior que 0.2
to_select_rows = [row for row in upper.index if any(upper.loc[row] > correlacao_minima)]

corrdf_grande = corrdf.loc[to_select_rows, to_select_columns]
corrdf_grande

# fig = px.imshow(corrdf_grande, title="Correlações entre diagnóstico de alzheimer com os dados maior que 0.2")
# fig.update_xaxes(side="top")
# fig.show()

sns.heatmap(corrdf_grande, annot=True).set(title=f"Correlações entre diagnóstico de alzheimer com os dados maior que {correlacao_minima}")
```

- Essas são as colunas com maior correlação com o diagnóstico de alzheimer, são todas colunas de Avaliações Cognitivas e Funcionais. Claro que problemas de memória (MemoryComplaints) e problemas comportamentais (BehavioralProblemas) iria ter uma relação positiva com alzheimer, e Pontuação do Mini-Exame do Estado Mental (MMSE) uma relação negativa, afinal quanto menor o valor do mini-exame, pior é seu estado mental. Avaliadores Funcionais (FuncionalAssessment) que é a determinação do nível de função e capacidade de uma pessoa para realizar tarefas diárias e requisitos de vida também tem uma relação negativa, quanto menor esse campo, maior prejuízo. E por último a pontuação de atividades de vida diária (ADL) que indica o nível de independência de uma pessoa na execução de tarefas relacionadas aos cuidados pessoais, quanto menor a pontuação maior o prejuízo também.
- Basicamente, todos esses dados são efeitos diversos do alzheimer.


```python
df2 = df.copy()[["Diagnosis", "EducationLevel", "FamilyHistoryAlzheimers", "Age"]]

df2.replace({
    "EducationLevel": {
        0: "Nenhuma", 
        1: "Ensino médio", 
        2: "Bacharelado", 
        3: "Pós-graduação"
    },
    "Diagnosis": {
        0: "Negativo",
        1: "Positivo",
    }
    }, 
    inplace=True)

df2_p = df2.groupby(['EducationLevel', 'Diagnosis']).size()
porcentagem = df2_p.groupby(level=0).apply(lambda x: x * 100 / sum(x)).values

df2_p = df2_p.reset_index().rename(columns={0: "Quantidade"})

df2_p["Porcentagem"] = porcentagem
df2_p["Porcentagem"] = df2_p["Porcentagem"].apply(lambda x: '{0:1.2f}%'.format(x))

fig = px.bar(
    df2_p, 
    title="casos de alzheimer e porcentagem por nível de educação",
    x="EducationLevel", 
    y="Quantidade", 
    color="Diagnosis",

    text="Porcentagem", 
    labels={
        "EducationLevel": "Nível de educação",
        "Diagnosis": "Diagnóstico"
    }
)
fig.update_xaxes(
    categoryorder='array', 
    categoryarray= ['Nenhuma', 'Ensino médio', 'Bacharelado', 'Pós-graduação']
)
```

- Levando em conta as porcentagens de alzheimer de acordo com o nível de educação, é perceptível uma pequena queda nos diagnóstico de alzheimer quanto mais avançado é a educação

```python
df2 = df.copy()[["Diagnosis", "FamilyHistoryAlzheimers"]]

df2.replace({
    "Diagnosis": {
        0: "Negativo",
        1: "Positivo",
    },
    "FamilyHistoryAlzheimers": {
        0: "Sem histórico",
        1: "Com histórico"
    }
    }, 
    inplace=True)

df2
df2_p = df2.groupby(['FamilyHistoryAlzheimers', 'Diagnosis']).size()
porcentagem = df2_p.groupby(level=0).apply(lambda x: x * 100 / sum(x)).values

df2_p = df2_p.reset_index().rename(columns={0: "Quantidade"})

df2_p["Porcentagem"] = porcentagem
df2_p["Porcentagem"] = df2_p["Porcentagem"].apply(lambda x: '{0:1.2f}%'.format(x))

fig = px.bar(
    df2_p, 
    x="FamilyHistoryAlzheimers", 
    y="Quantidade", 
    color="Diagnosis", 
    text="Porcentagem", 
    labels={
        "FamilyHistoryAlzheimers": "Histórico de familiar com alzheimer",
        "Diagnosis": "Diagnóstico"
    }
)
fig.show()
```

- É uma diferença pequena, mas aparentemente pessoas que tiveram familiares com alzheimer, tem menos chance de terem alzheimer, isso indica provavelmente, que alguém que teve um familiar com alzheimer, tome prevenções e cuidados para não ter a mesma doença

```python
df2 = df.replace({
        "Diagnosis": {
            0: "Negativo",
            1: "Positivo",
        }
    })

colunas = {
    "PhysicalActivity": "Atividade física", 
    "Age": "Idade", 
    "Smoking": "Fumo", 
    "AlcoholConsumption": "Consumo de Alcool", 
    "DietQuality": "Qualidade de dieta"
}
for coluna in colunas:
    fig = px.histogram(
        df2,
        x=coluna, 
        color="Diagnosis", 
        barnorm="percent",
        #text_auto=True
    )
    fig.update_layout(
        xaxis_title=colunas[coluna], 
        yaxis_title="Porcentagem (%)",
        title=f"Porcentagem de diagnóstico de alzheimer por {colunas[coluna].lower()}"
    )
    fig.show()
```
